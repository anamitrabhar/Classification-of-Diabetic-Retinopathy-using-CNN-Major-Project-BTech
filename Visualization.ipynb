```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
chk = torch.load('multiclass classification with state_dict.pt')
```

```python
for i in chk:
    print(i)
```

```python
trad_t_loss=chk['training_losses']
trad_v_loss=chk['valid_losses']
trad_acc = chk['accuracies']
trad_c_acc = chk['classes_accuracies']
trad_confusion_matrix = chk['confusion_matrix']
trad_y_pred = chk['y_pred']
trad_y_true = chk['y_true']
print(len(trad_t_loss))
```

```python
trad_epochs = list(range(1,len(trad_t_loss)+1))
```

```python
print(trad_epochs)
```

```python
x = trad_epochs
y = trad_t_loss
z = trad_v_loss
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss  v/s Epoch' )
plt.plot(x,y,label = 'Training Loss')
plt.plot(x,z,label = 'Validation Loss')
plt.legend()
plt.show()
```

```python
x = trad_epochs
y2 = trad_acc

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy  v/s Epoch')
plt.plot(x,y2)


plt.show()
```

```python
trad_classes = ['Level 0', 'Level 1', 'Level 2', 'Level 3', 'Level 4']


plt.barh(trad_classes,trad_c_acc,align = 'center')
plt.xlabel('Accuracy (%)')
plt.ylabel('Class')
plt.title('Class Accuracies')
plt.show()
```

```python
sns.heatmap(trad_confusion_matrix, annot=True , cmap='Blues' ,fmt='g')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, jaccard_score, roc_auc_score
from numpy import array
from numpy import argmax
#from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix
```

```python
v_labels=trad_y_pred
v_pred=trad_y_true
```

```python
precision = precision_score(v_labels,v_pred,average='weighted')
recall = recall_score(v_labels,v_pred,average='weighted')
f1 = f1_score(v_labels,v_pred,average='weighted')
jaccard = jaccard_score(v_labels,v_pred,average='weighted')
```

```python
print("precision = ", precision)
print("recall = ",recall)
print("f1 = ",f1)
print("jaccard = ",jaccard)
```
